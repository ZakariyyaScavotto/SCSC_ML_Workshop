Columns details:
* Track - Title of the song
* Artist - Name of the performing artist/band
* Year - The year the track was released
* Duration - Length of the track in milliseconds
* Time Signature - Musical time signature of the track = How many beats are in each bar
* Danceability - How suitable the track is for dancing, values from 0.0 to 1.0
* Energy - Measure of the intensity of the track, values from 0.0 to 1.0
* Key - The key of the track, eg: 0 = C
* Loudness - The overall loudness of the track in decibels, dB.
* Mode - The specific type of scale used, major (1) or minor (0)
* Speechiness - Measure of the presence of spoken words or vocal activity in a track rather than musical sounds
* Acousticness - How acoustic is the track, 0.0 to 1.0
* Instrumentalness - Measure of the likelihood that the track is instrumental, ranging from 0.0 to 1.0.
* Liveness - measure of the presence of a live audience in the track, ranging from 0.0 to 1.0
* Valence - measures the emotional quality of a track, specifically how positive or negative it sounds, value between 0.0 and 1.0
* Tempo - Speed of the track in beats per minute (BPM)
* Popularity - measure of the track's popularity, ranging from 0 to 100
* Genre - Pop, Metal, EDM, Rap etc

```python
# Group by year and count the number of tracks
tracks_per_year = df.groupby('Year').size().sort_index()

# Plot number of tracks per year
plt.figure(figsize=(12,6))
sns.lineplot(x=tracks_per_year.index, y=tracks_per_year.values)
plt.title('Number of Tracks Released Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Tracks')
plt.show()
```

```python
# Find the earliest year + # tracks in the dataset
earliest_year = df['Year'].min()
tracks_earliest_year = df[df['Year'] == earliest_year].shape[0]
print(f"The earliest year in the dataset is {earliest_year} with {tracks_earliest_year} tracks")

# Find the most recent year + # tracks in the dataset
most_recent_year = df['Year'].max()
tracks_most_recent_year = df[df['Year'] == most_recent_year].shape[0]
print(f"The most recent year in the dataset is {most_recent_year} with {tracks_most_recent_year} tracks")
```

```python
print(df[['Duration']])
# Create a temporary column for duration in minutes
df_temp = df.copy()
df_temp['Duration_min'] = df_temp['Duration'] / 60000
```

```python
# Create a temporary column for duration in minutes
df_temp = df.copy()
df_temp['Duration_min'] = df_temp['Duration'] / 60000

# Plot duration distribution over time
plt.figure(figsize=(12,6))
sns.lineplot(x='Year', y='Duration_min', data=df_temp, estimator='mean')
# automatic confidence interval in lineplot
plt.title('Average Track Duration Over Time')
plt.xlabel('Year')
plt.ylabel('Average Duration (minutes)')
plt.show()
```

```python
# Valence, Danceability and Energy over time
plt.figure(figsize=(12,6))
sns.lineplot(x='Year', y='Danceability', data=df, estimator='mean', label='Danceability')
sns.lineplot(x='Year', y='Energy', data=df, estimator='mean', label='Energy')
sns.lineplot(x='Year', y='Valence', data=df, estimator='mean', label='Valence')
plt.title('Valence, Danceability and Energy Over Time')
plt.xlabel('Year')
plt.ylabel('Average Value')
plt.legend()
plt.show()

# Speechiness and Instrumentalness over time
plt.figure(figsize=(12,6))
sns.lineplot(x='Year', y='Speechiness', data=df, estimator='mean', label='Speechiness')
sns.lineplot(x='Year', y='Instrumentalness', data=df, estimator='mean', label='Instrumentalness')
plt.title('Speechiness and Instrumentalness Over Time')
plt.xlabel('Year')
plt.ylabel('Average Value')
plt.legend()
plt.show()

# Acousticness and Liveness over time
plt.figure(figsize=(12,6))
sns.lineplot(x='Year', y='Acousticness', data=df, estimator='mean', label='Acousticness')
sns.lineplot(x='Year', y='Liveness', data=df, estimator='mean', label='Liveness')
plt.title('Acousticness and Liveness Over Time')
plt.xlabel('Year')
plt.ylabel('Average Value')
plt.legend()
plt.show()
```

```python
# Loudness over time
plt.figure(figsize=(12,6))
sns.lineplot(x='Year', y='Loudness', data=df, estimator='mean')
plt.title('Average Track Loudness Over Time')
plt.xlabel('Year')
plt.ylabel('Loudness (dB)')
plt.show()
```

```python
# create 2 boxplot of 1990 and 2024 for tempo
plt.figure(figsize=(12,6))
sns.boxplot(x='Year', y='Tempo', data=df[(df['Year'] == 1990) | (df['Year'] == 2024)])
plt.title('Tempo Distribution in 1990 and 2024')
plt.xlabel('Year')
plt.ylabel('Tempo')
```

```python
# Set figure size
plt.figure(figsize=(12, 6))

# Create a box plot to show the distribution of tempo over time
sns.boxplot(x='Year', y='Tempo', data=df)

# Set plot title and labels
plt.title('Distribution of Tempo Over Time')
plt.xlabel('Year')
plt.ylabel('Tempo (BPM)')

# Show the plot
plt.xticks(rotation=90)
plt.show()
```

```python
# clean outliers
# Find and display rows with Tempo value of zero
rows_with_zero_tempo = df[df['Tempo'] == 0]
print("Rows with Tempo value of zero:")
print(rows_with_zero_tempo)

# Remove rows with Tempo value of zero
df = df[df['Tempo'] != 0]
df
```

```python
# Set figure size
plt.figure(figsize=(12, 6))

# Create a box plot to show the distribution of tempo over time
sns.boxplot(x='Year', y='Tempo', data=df)

# Set plot title and labels
plt.title('Distribution of Tempo Over Time')
plt.xlabel('Year')
plt.ylabel('Tempo (BPM)')

# Show the plot
plt.xticks(rotation=90)
plt.show()
```

```python
# Grouping by year and genre to get the count of tracks per genre over time
genre_evolution = df.groupby(['Year', 'Genre']).size().unstack().fillna(0)

# Plotting the evolution of genres over time
genre_evolution.plot(figsize=(12, 6), cmap='tab20')
plt.title('Genre Evolution Over Time')
plt.ylabel('Number of Tracks')
plt.xlabel('Year')
plt.show()
```

```python
# Group by Genre and count the number of tracks in each genre
genre_counts = df['Genre'].value_counts()

# Plot a bar chart for genres sorted by track count
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values)
plt.title('Number of Tracks per Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Tracks')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Create a scatter plot for loudness vs. danceability
sns.scatterplot(x='Loudness', y='Danceability', data=df, alpha=0.5)

plt.title('Loudness vs. Danceability')
plt.xlabel('Loudness (dB)')
plt.ylabel('Danceability')
plt.show()
```

```python
plt.figure(figsize=(14, 7))

# Calculate average popularity for each genre
average_popularity = df.groupby('Genre')['Popularity'].mean().sort_values()

# Plot the average popularity
sns.barplot(x=average_popularity.index, y=average_popularity.values)

plt.title('Average Popularity by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Popularity')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# can look at correlation between Valance, Danceability, and Energy
# Calculate correlation matrix
correlation = 1 - df[['Valence', 'Danceability', 'Energy']].corr()

# Display the correlation matrix
correlation
```

```python
# display the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='plasma', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(12,8))
correlation_matrix = df[['Danceability', 'Energy', 'Loudness', 'Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo', 'Popularity']].corr()

# Heatmap of the correlations
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Musical Features')
plt.show()
```

```python
highest_popularity_tracks = df.loc[df.groupby('Genre')['Popularity'].idxmax()]
print(highest_popularity_tracks[['Track','Genre','Popularity']])

plt.figure(figsize=(10, 6))
bars = plt.barh(highest_popularity_tracks['Track'], highest_popularity_tracks['Popularity'], color='skyblue')
plt.xlabel('Popularity')
plt.ylabel('Track')
plt.title('Track Popularity by Genre')

# Add genre labels to the chart
for bar, genre in zip(bars, highest_popularity_tracks['Genre']):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, genre, va='center')

plt.tight_layout()
plt.show()
```
